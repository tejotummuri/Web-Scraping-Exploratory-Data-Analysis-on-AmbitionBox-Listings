```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import time

Name = [] 

location = []

Ratings = []

pagenum = []

Reviews = []

Salaries = []

Interviews = []
jobs = []
Benifits = []
Awards = []

total_time = time.time()
for i in range(1, 11): 

    start_time = time.time()

   
    url = f"https://www.ambitionbox.com/companies-in-hyderabad?page={i}"

    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)

    

    soup = BeautifulSoup(response.text, "html.parser")

    for x in soup.find_all('div', attrs={'class':'companyCardWrapper'}):

        name_tag = x.find('a', attrs={"class":"companyCardWrapper__companyName"})

        location_tag = x.find('span', attrs={'class':'companyCardWrapper__interLinking'})
        
        rat_tag = x.find('div', attrs={'class':'companyCardWrapper__companyRatingWrapper'})   # discounted price

        reviews_tag = x.find('span', attrs={'class':'companyCardWrapper__ActionCount'})

        salaries_tag = x.find('span', attrs={'class':'companyCardWrapper__ActionCount'})

        interviews_tag = x.find('span',attrs={'class':'companyCardWrapper__ActionCount'})
        job_tag = x.find('span',attrs={'class':'companyCardWrapper__ActionCount'})
        benifits_tag =  x.find('span',attrs={'class':'companyCardWrapper__ActionCount'})
        award_tag = x.find('div', attrs = {'class':"award-leaf"})
        

        Name.append(name_tag.text.strip() if name_tag else np.nan)

        Ratings.append(rat_tag.text.strip() if rat_tag else np.nan)

        location.append(location_tag.text.strip() if location_tag else np.nan)
        
        Awards.append(award_tag.text.strip() if award_tag else np.nan)
        action_counts = x.find_all('span', class_='companyCardWrapper__ActionCount')

        # Safe extraction (some cards may not have all 5)
        reviews = action_counts[0].text.strip() if len(action_counts) > 0 else np.nan
        salaries = action_counts[1].text.strip() if len(action_counts) > 1 else np.nan
        interviews = action_counts[2].text.strip() if len(action_counts) > 2 else np.nan
        job_count = action_counts[3].text.strip() if len(action_counts) > 3 else np.nan
        benefits = action_counts[4].text.strip() if len(action_counts) > 4 else np.nan

        Reviews.append(reviews)
        Salaries.append(salaries)
        Interviews.append(interviews)
        jobs.append(job_count)
        Benifits.append(benefits)

        pagenum.append(i)
 
    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time)) 



data = pd.DataFrame({
    "Name": Name,
    "Ratings": Ratings,
    "Location": location,
    "Reviews": Reviews,
    "Salaries": Salaries,
    "Interviews": Interviews,
    "jobs":jobs,
    "Benifits":Benifits,
    "Awards":Awards,
     "Page Number": pagenum})
```

```python
import pandas as pd

def convert_value(x):
    x = str(x).lower()
    
    if 'k' in x:
        return float(x.replace('k','')) * 1000
    elif 'l' in x:
        return float(x.replace('l','')) * 100000
    else:
        return float(x)

cols = ['Reviews','Salaries','Interviews','jobs','Benifits']

for c in cols:
    data[c] = data[c].apply(convert_value)
```
